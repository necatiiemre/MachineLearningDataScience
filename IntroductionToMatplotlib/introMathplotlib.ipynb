## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupiter
#%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot Data
ax.plot(x, y)

# 4. customize plot
ax.set_title("Simple Plot")
ax.set(xlabel="x-axis",
       ylabel = "y-axis")

# 5.  Save & Show (You save the whole figure)
fig.savefig("sample.png")
```

### Two options for subplots 

```python
# Subplot option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize = (10 , 5))
# Plot to each different axis
ax1.plot(x , x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
ax4.hist(np.random.rand(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize = (10, 5))
# Plot to each different index
ax[0, 0].plot(x , x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
ax[1, 1].hist(np.random.rand(1000));
```

### Which one should you use ? (pyplot vs matplotlib OO method )
* When plotting something quickly, okay to use the pyplot method
* When plotting something mode advanced, use the OO method



```python
over_50 = heard_disease[heard_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y = "chol",
             c = "target");
```

```python
## OO Method mixed with pyplot method
fig, ax = plt.subplots(figsize = (10,10))
ax.set_xlim([45, 100])
over_50.plot(kind="scatter",
             x="age",
             y = "chol",
             c = "target",
             ax=ax);


```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y= over_50["chol"],
                     c = over_50["target"]);

# Customize the plot
ax.set(title="Heard Disease and Cholosterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Add Horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle ="--");
```

```python
over_50.head(10)
```

```python
# Subplot od chol, age, thalach

fig, (ax0, ax1) = plt.subplots(figsize = (10, 10),ncols=1,nrows=2, sharex=True) # You can not to write x 

# Plot the data
scatter = ax0.scatter(x = over_50["age"],
                     y= over_50["chol"],
                     c = over_50["target"]);

# Customize the plot
ax0.set(title="Heard Disease and Cholosterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend
ax0.legend(*scatter.legend_elements(), title = "Target");

# Add Horizontal line
ax0.axhline(over_50["chol"].mean(),
           linestyle ="--");


# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                     y= over_50["thalach"],
                     c = over_50["target"])

# Customize the plot
ax1.set(title="Heard Disease and Max Heart Rate",
       xlabel = "Age",
       ylabel = "Max Heart Rate");

# Add a legend
ax1.legend(*scatter.legend_elements(), title = "Target");

# Add Horizontal line
ax1.axhline(over_50["thalach"].mean(),
           linestyle ="--");

fig.suptitle("Heard Disease Analysis", fontsize = 16, fontweight = "bold");
```
