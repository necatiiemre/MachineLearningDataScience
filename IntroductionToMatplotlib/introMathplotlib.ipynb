## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupiter
#%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot Data
ax.plot(x, y)

# 4. customize plot
ax.set_title("Simple Plot")
ax.set(xlabel="x-axis",
       ylabel = "y-axis")

# 5.  Save & Show (You save the whole figure)
fig.savefig("sample.png")
```

## Making Figures with NumPy Arrays

We want: 
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# Create some Data
x = np.linspace(0, 10 , 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots(figsize= (10 , 10))
ax.plot(x, x**2)
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter" : 10,
                     "Peanut butter" : 8,
                     "Cashew butter" : 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Emre 's Nut Butter Store ",  ylabel="Price ($)", xlabel="Type")
```

```python
fig ,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some data for Histograms and plot it
x = np.random.randn(1000) 
```

```python
x[:10]
```

```python
fig, ax = plt.subplots()
ax.hist(x)
```

### Two options for subplots 

```python
# Subplot option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize = (10 , 5))
# Plot to each different axis
ax1.plot(x , x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
ax4.hist(np.random.rand(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize = (10, 5))
# Plot to each different index
ax[0, 0].plot(x , x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
ax[1, 1].hist(np.random.rand(1000));
```
